```python
def get_data_overview(data):
    # Get column info
    num_data = data.select_dtypes(include=['float64', 'int64'])
    cat_data = data.select_dtypes(include=['object'])
    
    num_info = pd.DataFrame(num_data.dtypes.astype(str), columns=["Data Type"])
    num_info.index.name = "Column Name"
    num_info["No of Unique Values"] = num_data.nunique()
    num_info["No of Missing Values"] = num_data.isnull().sum()
    num_info["% of Missing Values"] = round((num_info["No of Missing Values"] / len(num_data)) * 100, 2)
    num_info["No of Duplicates"] = num_data.duplicated().sum()
    
    cat_info = pd.DataFrame(cat_data.dtypes.astype(str), columns=["Data Type"])
    cat_info.index.name = "Column Name"
    cat_info["No of Unique Values"] = cat_data.nunique()
    cat_info["No of Missing Values"] = cat_data.isnull().sum()
    cat_info["% of Missing Values"] = round((cat_info["No of Missing Values"] / len(cat_data)) * 100, 2)
    cat_info["No of Duplicates"] = cat_data.duplicated().sum()

    # Print overview table
    print(f"{'-'*120}\n{'Numeric Columns':^120}\n{'-'*120}\n{num_info.index.name:20} {'Data Type':10} {'No of Unique Values':20} {'No of Missing Values':20} {'% of Missing Values':20} {'No of Duplicates':20}\n{'-'*120}")
    for i in range(len(num_info)):
        print(f"{num_info.index[i]:20} {str(num_info.iloc[i, 0]):>10} {num_info.iloc[i, 1]:>20} {num_info.iloc[i, 2]:>20} {num_info.iloc[i, 3]:>20} {num_info.iloc[i, 4]:>20}")
        
    print(f"{'-'*120}\n{'Categorical Columns':^120}\n{'-'*120}\n{cat_info.index.name:20} {'Data Type':10} {'No of Unique Values':20} {'No of Missing Values':20} {'% of Missing Values':20} {'No of Duplicates':20}\n{'-'*120}")
    for i in range(len(cat_info)):
        print(f"{cat_info.index[i]:20} {str(cat_info.iloc[i, 0]):>10} {cat_info.iloc[i, 1]:>20} {cat_info.iloc[i, 2]:>20} {cat_info.iloc[i, 3]:>20} {cat_info.iloc[i, 4]:>20}")
```

```python
import pandas as pd

def get_five_number_summary(data):
    num_data = data.select_dtypes(include=['float64', 'int64'])
    stats = num_data.describe().transpose()[['min', '25%', '50%', '75%', 'max']]
    stats['Range'] = stats['max'] - stats['min']
    stats.rename(columns={'50%': 'Median'}, inplace=True)

    # Print summary table
    print(f"{'-'*85}\n{'Column Name':20} {'Minimum':>10} {'1st Quartile':>15} {'Median':>15} {'3rd Quartile':>15} {'Maximum':>10} {'Range':>10}\n{'-'*85}")
    for i in range(len(stats)):
        print(f"{stats.index[i]:20} {stats.iloc[i, 0]:>10.2f} {stats.iloc[i, 1]:>15.2f} {stats.iloc[i, 2]:>15.2f} {stats.iloc[i, 3]:>15.2f} {stats.iloc[i, 4]:>10.2f} {stats.iloc[i, 5]:>10.2f}")
```

### Clean and preprocess the data, handle missing values and outliers

```python
def clean_data(data):
    # Drop duplicates
    data.drop_duplicates(inplace=True)
    
    # Convert columns to numeric
    for col in data.columns:
        try:
            data[col] = pd.to_numeric(data[col])
        except:
            pass
    
    # Remove columns with > 60% missing values
    missing_values = data.isnull().mean()
    columns_to_drop = list(missing_values[missing_values > 0.6].index)
    data.drop(columns_to_drop, axis=1, inplace=True)
    
    # Remove rows with missing values
    data.dropna(inplace=True)
    
    # Remove outliers (assuming all numerical columns)
    num_data = data.select_dtypes(include=['float64', 'int64'])
    q1 = num_data.quantile(0.25)
    q3 = num_data.quantile(0.75)
    iqr = q3 - q1
    num_data = num_data[~((num_data < (q1 - 1.5 * iqr)) | (num_data > (q3 + 1.5 * iqr))).any(axis=1)]
    data = pd.concat([num_data, data.select_dtypes(include=['object'])], axis=1)
    
    # Impute missing values
    num_cols = data.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = data.select_dtypes(include=['object']).columns.tolist()
    for col in num_cols:
        data[col].fillna(data[col].mean(), inplace=True)
    for col in cat_cols:
        data[col].fillna(data[col].mode()[0], inplace=True)
        
        
    return data
```

```python
def convert_to_categorical(data):
   
    # Get column info
    num_data = data.select_dtypes(include=['float64', 'int64'])
    info = pd.DataFrame(num_data.dtypes.astype(str), columns=["Data Type"])
    info.index.name = "Column Name"
    info["No of Unique Values"] = num_data.nunique()

    # Identify columns with less than 20 unique values or only one unique value
    categorical_cols = []
    for col in info.index:
        if info.loc[col, "No of Unique Values"] <= 1:
            data = data.drop(col, axis=1)
        elif info.loc[col, "No of Unique Values"] < 20:
            categorical_cols.append(col)

    # Convert identified columns to categorical variables
    for col in categorical_cols:
        data[col] = data[col].astype("category")

    return data
```

```python
import seaborn as sns

def plot_heatmap(df, fig_size=(20, 12)):
    # Select only numerical columns
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

    # Calculate the correlation matrix
    corr_matrix = df[numerical_cols].corr()

    # Set the figure size
    sns.set(rc={'figure.figsize': fig_size})

    # Plot the heatmap
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
def bar_graph(data):
    # Create a copy of the original data
    data_copy = data.copy()
    
    # Get categorical columns
    cat_cols = data_copy.select_dtypes(include=['object']).columns
    
    # Loop through each categorical column
    for col in cat_cols:
        # Get the value counts
        counts = data_copy[col].value_counts()
        
        # If the column has more than 10 categories, get the top 10
        if len(counts) > 10:
            counts = counts.head(10)
        
        # Plot the bar graph
        counts.plot(kind='bar')
        plt.title(col)
        plt.show()

# Call the function to plot the bar graph
bar_graph(df)
```
